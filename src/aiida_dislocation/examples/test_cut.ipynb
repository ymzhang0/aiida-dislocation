# Visualize A1 structure

```python
ase_atom = read('./structures/cif/Al-225-A1.cif', format='cif')

P = numpy.array([
    [10, 0, 0],
    [0, 10, 0],
    [0, 0, 10],
])

slipping_systems = {
    '100': {
        'cut_plane': [1, 1, -1],
        'a': [-1, 1, 1],
        'b': [1, -1, 1],
    },
    '110': {
        'cut_plane': [1, 0, 0],
        'a': [0, 1, -1],
        'b': [-1, 1, 1],
    },
    '111': {
        'cut_plane': [1, 1, 1],
        'a': [0, 1, -1],
        'b': [-2, 1, 1],
        },
    }

slipping_system = slipping_systems['111']

grouped_by_plane = group_by_plane(ase_atom, P, slipping_system['cut_plane'])

from matplotlib import pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
        

l = list(grouped_by_plane.items())
fig, ax = plt.subplots()

plot_layer(ase_atom, l[9][1], ax, slipping_system['a'], slipping_system['b'], 'red', 'Al')
plot_layer(ase_atom, l[10][1], ax, slipping_system['a'], slipping_system['b'], 'blue', 'Al')
plot_layer(ase_atom, l[11][1], ax, slipping_system['a'], slipping_system['b'], 'black', 'Al')
# plot_layer(l[6][1], ax, a, b, 'black', 'Al')
# plot_layer(l[7][1], ax, a, b, 'blue', 'Al')
# plot_layer(l[8][1], ax, a, b, 'cyan', 'Al')
plt.show()




```
